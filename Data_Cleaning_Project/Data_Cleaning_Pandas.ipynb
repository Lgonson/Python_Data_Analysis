```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\lgons\workspace\Python\Data_Cleaning_Project\Customer_Call_List.xlsx')
```

```python
df.head()
```

```python
#drop duplicates
df = df.drop_duplicates()
```

```python
#Drop unnecessary columns
df = df.drop(columns = 'Not_Useful_Column')
```

```python
#Cleaning Last_Name column
df['Last_Name'] = df['Last_Name'].str.strip('./_123')
```

```python
#Cleaning Phone Number column
#delete all values except of a to z, A to Z, and 0 to 9
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
#Convert all numbers into string to handle them
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:str(x)) 
#Separate the numbers with -
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:9])  
#Delete nan-- and 'Na--
df['Phone_Number'] = df['Phone_Number'].replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].replace('Na--', '')
```

```python
#Cleaning Address column
df[['StreetAdress', 'State', 'ZIP']] = df['Address'].str.split(',', n=2, expand = True)
df = df.drop("Address", axis = 1)
```

```python
#cleaning Paying Customer column

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

df['Paying Customer'] = df['Paying Customer'].str.replace('Y', 'Yes')
df['Paying Customer'] = df['Paying Customer'].str.replace('N', 'No')
```

```python
#cleaning Do_Not_Contact column

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
#replace NaN, No/a, N/a
df.replace('N/a', '', inplace = True)
df.replace('No/a', '', inplace = True)
df.replace('NaN', '', inplace = True)
```
